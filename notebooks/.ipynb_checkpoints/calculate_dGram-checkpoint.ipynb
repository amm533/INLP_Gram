### LDA (Linear Discriminant Analysis)

```python
m_words = []
f_words = []

gram_emb_set_m = []
gram_emb_set_f = []
    
with open("../data/sets_palabras/ES/gram/ES_gram_LDA_m.txt", "r", encoding="utf-8") as f: 
    for line in f:
        word = line.strip()
    
        if word in m_words:
            continue
                
        if word not in biased_model:
            continue
                    
        m_words.append(word)
        gram_emb_set_m.append(biased_model[word])

with open("../data/sets_palabras/ES/gram/ES_gram_LDA_f.txt", "r", encoding="utf-8") as f: 
    for line in f:
        word = line.strip()
    
        if word in f_words:
            continue
                
        if word not in biased_model:
            continue        
                
        f_words.append(word)
        gram_emb_set_f.append(biased_model[word])

gram_emb_set_f = np.array(gram_emb_set_f)
gram_emb_set_m = np.array(gram_emb_set_m)
print(gram_emb_set_m.shape)
print(gram_emb_set_f.shape)
```

```python
import directions

dGram = directions.get_gender_direction_LDA_emb(gram_emb_set_f, gram_emb_set_m)
```

```python
# alternative
dGram = directions.get_gender_direction_LDA(biased_model, "../data/sets_palabras/ES/gram/ES_gram_LDA_f.txt"
                                                        , "../data/sets_palabras/ES/gram/ES_gram_LDA_m.txt")
```

```python
np.savetxt("../data/directions/dGram", dGram)
```
