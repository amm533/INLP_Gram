```python
import sys
sys.path.append("../src")
import utils

from gensim.models.keyedvectors import KeyedVectors
import numpy as np
```

```python
ES_biased_model = KeyedVectors.load('../data/embeddings/keyedvectors/model_esp.kv', mmap='r')
```

```python
ES_inlp_debiased_model = KeyedVectors.load('../data/embeddings/keyedvectors/ES_inlp_debiased_model.kv', mmap='r')
```

```python
inlpES_debiased_model = KeyedVectors.load('../data/embeddings/keyedvectors/inlpES_debiased_model.kv', mmap='r')
```

```python
masculine_words = []
feminine_words = []

count = 0

with open("../data/sets_palabras/ES/gram/ES_GGCTest_m.txt", "r", encoding='utf-8') as f:
    for line in f:
        count = count +1
        if(count < 100):
            word = line.strip()
            if word not in masculine_words:
                masculine_words.append(word)

count = 0
with open("../data/sets_palabras/ES/gram/ES_GGCTest_f.txt", "r", encoding='utf-8') as f:
    for line in f:
        count = count +1
        if(count < 100):
            word = line.strip()
            if word not in feminine_words:
               feminine_words.append(word)

grammar_pairs = []
for f,m in zip(feminine_words, masculine_words):
    pair = [f,m]
    grammar_pairs.append(pair)

print(len(grammar_pairs))
```
